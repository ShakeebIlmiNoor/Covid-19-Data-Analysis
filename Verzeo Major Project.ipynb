## Importing Dataset

```python
# Importing pandas for performing operations on dataset
import pandas as pd

# Importing dataset which is in the form of a CSV File
dataset = pd.read_csv("owid-covid-data.csv")
```

```python
# Printing first 10 rows
dataset.head(10)
```

## Subsetting those rows which have India in their location

```python
data_ind = dataset[dataset["location"]=='India']
```

```python
# Getting dimensions of the new dataset
data_ind.shape
```

```python
# Re-indexing the Dataset
data_ind.index=list(range(1,220))

# Finding information of null values in Dataset
data_ind.info()
```

## Handling Missing Values

```python
# Filling the null values of continuous numeric with the mean of the column
data_ind.fillna(data_ind.mean(),inplace=True)

# Filling the null values of the categorical column with the mode of the column
data_ind['tests_units'].fillna("samples tested", inplace = True)

# Finding information of the null values in the Dataset
data_ind.info()
```

## Converting Date Column to Ordinal

```python
# Importing datetime and converting date column to ordinal form
import datetime as dt
data_ind.loc[:,'date'] = pd.to_datetime(data_ind['date'])
data_ind.loc[:,'date'] = data_ind['date'].map(dt.datetime.toordinal)
```

## Dropping Useless Categorical Columns

```python
# Dropping all columns which are useless for modelling purposes
useful_dat = data_ind.drop(['iso_code','continent','location','tests_units'], axis = 1)
```

```python
useful_dat.info()
```

## Selecting "total_cases" Column as Target Column

```python
y = useful_dat['total_cases']
```

## Selecting Other Columns as Features

```python
X = useful_dat.drop('total_cases', axis = 1)
```

## Performing Train-Test Split

```python
# Implementing Train-test split
from sklearn.model_selection import train_test_split
# We perform hyperparameter tuning during Train-test split and choose the optimum values to provide maximum accuracy
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 64)
```

## Data Modelling

### Implementing Linear Regression

```python
from sklearn.linear_model import LinearRegression
# Instantiation
linreg = LinearRegression()

# Fitting the model
linreg.fit(X_train, y_train)
```

### Printing Accuracy of Linear Regression Model

```python
accuracy = linreg.score(X_test, y_test) * 100
print("Accuracy of the model while using Linear Regression is " + str(accuracy) + " %")
```

### Implementing Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
# Instantiation
random = RandomForestRegressor()

# Fitting the model
random.fit(X_train, y_train)
```

### Printing Accuracy of Random Forest Regressor

```python
accuracy = random.score(X_test, y_test) * 100
print("Accuracy of the model while using Random Forest Regressor is " + str(accuracy) + " %")
```
